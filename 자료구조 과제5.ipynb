```python
class Heap:
	def __init__(self, *args):
		if len(args) != 0:
			self.__A = args[0] # 파라미터로 온 리스트
		else:
			self.__A = []

	# [알고리즘 8-2] 구현: 힙에 원소 삽입하기(재귀 알고리즘 버전)
	def insert(self, x):
		self.__A.append(x)
		self.__percolateUp(len(self.__A)-1)

	# 스며오르기
	def __percolateUp(self, i:int):
		parent = (i - 1) // 2
		if i > 0 and self.__A[i] > self.__A[parent]:
			self.__A[i], self.__A[parent] = self.__A[parent], self.__A[i]
			self.__percolateUp(parent)

	# [알고리즘 8-2] 구현: 힙에서 원소 삭제하기
	def deleteMax(self):
		# heap is in self.__A[0...len(self.__A)-1]
		if (not self.isEmpty()):
			max = self.__A[0]
			self.__A[0] = self.__A.pop() # *.pop(): 리스트의 끝원소 삭제 후 원소 리턴
			self.__percolateDown(0)
			return max
		else:
			return None

	# 스며내리기
	def __percolateDown(self, i:int):
		# Percolate down w/ self.__A[i] as the root
		child = 2 * i + 1  # left child
		right = 2 * i + 2  # right child
		if (child <= len(self.__A)-1):
			if (right <= len(self.__A)-1 and self.__A[child] < self.__A[right]):
				child = right  # index of larger child
			if self.__A[i] < self.__A[child]:
				self.__A[i], self.__A[child] = self.__A[child], self.__A[i]
				self.__percolateDown(child)

	def max(self):
		return self.__A[0]

	# 힙 만들기
	def buildHeap(self):
		for i in range((len(self.__A) - 2) // 2, -1, -1):
			self.__percolateDown(i)

	# 힙이 비었는지 확인하기
	def isEmpty(self) -> bool:
		return len(self.__A) == 0

	def clear(self):
		self.__A = []

	def size(self) -> int:
		return len(self.__A)


h= Heap()
h.clear()

# 생일 데이터
h.insert(20031015)  # 강다원
h.insert(20051214)  # 강민주
h.insert(20041103)  # 강수아
h.insert(20041108)  # 강윤서
h.insert(20041228)  # 강주영
h.insert(20041004)  # 권보은
h.insert(20030805)  # 김나림
h.insert(20051005)  # 김나영
h.insert(20040203)  # 김나현
h.insert(20040117)  # 김난영
h.insert(20000209)  # 김남은
h.insert(20030304)  # 김다연
h.insert(20050923)  # 김도경
h.insert(20051002)  # 김명신
h.insert(20050324)  # 김미배
h.insert(20051202)  # 김민경
h.insert(20040210)  # 김민영
h.insert(20040517)  # 김민주
h.insert(20041026)  # 김민주
h.insert(20020911)  # 김보민
h.insert(20040628)  # 김서빈
h.insert(20050203)  # 김소민
h.insert(20030110)  # 김수련
h.insert(20041207)  # 김수민
h.insert(20030124)  # 김승연
h.insert(20030910)  # 김시연
h.insert(20031128)  # 김아린
h.insert(20051031)  # 김여원
h.insert(20010826)  # 김연진
h.insert(20051019)  # 김예빈
h.insert(20040412)  # 김예원
h.insert(20050422)  # 김정민
h.insert(20000922)  # 김주영
h.insert(20030110)  # 김주원
h.insert(20050417)  # 김주하
h.insert(20040722)  # 김지우
h.insert(20050123)  # 김채린
h.insert(20050910)  # 김채민
h.insert(20040409)  # 김채현
h.insert(20030212)  # 김하린
h.insert(20051001)  # 김혜인
h.insert(20050501)  # 김혜정
h.insert(20011212)  # 김효리
h.insert(20041104)  # 나주희
h.insert(20050325)  # 남궁수아
h.insert(20050316)  # 노은서
h.insert(20041105)  # 두경은
h.insert(20050214)  # 민고은
h.insert(20040428)  # 박서연
h.insert(20040514)  # 박성연
h.insert(20040728)  # 박지호
h.insert(20000507)  # 박찬미
h.insert(20030603)  # 박혜린
h.insert(20051102)  # 배시은
h.insert(20040802)  # 변수연
h.insert(20051024)  # 서홍빈
h.insert(20030122)  # 선예영
h.insert(20030104)  # 성유빈
h.insert(20050620)  # 손지원
h.insert(20041108)  # 송민서
h.insert(20041206)  # 신다연
h.insert(20051230)  # 신수민
h.insert(20040707)  # 신지예
h.insert(20050126)  # 신희영
h.insert(20040420)  # 안소민
h.insert(20031022)  # 안소형
h.insert(20030603)  # 안수민
h.insert(19851205)  # 안양준
h.insert(20040501)  # 안정민
h.insert(20030926)  # 안희랑
h.insert(20030702)  # 양소윤
h.insert(20051009)  # 여지혜
h.insert(20020505)  # 오다현
h.insert(20050328)  # 오세은
h.insert(20050712)  # 오예준
h.insert(20050509)  # 오하민
h.insert(20050920)  # 유채원
h.insert(20050519)  # 윤서영
h.insert(20040413)  # 윤소정
h.insert(20030206)  # 윤여빈
h.insert(20050517)  # 윤혜진
h.insert(20040503)  # 은유빈
h.insert(20040927)  # 이가연
h.insert(20041007)  # 이서연
h.insert(20051112)  # 이서영
h.insert(20051225)  # 이서영
h.insert(20050305)  # 이서진
h.insert(20051028)  # 이서진
h.insert(20040609)  # 이서현
h.insert(20040118)  # 이세윤
h.insert(20040910)  # 이수빈
h.insert(20041005)  # 이승주
h.insert(20010904)  # 이아현
h.insert(20040316)  # 이예린
h.insert(20021215)  # 이예림
h.insert(20030109)  # 이예은
h.insert(20020324)  # 이우정
h.insert(20050602)  # 이원진
h.insert(20050601)  # 이유빈
h.insert(20041107)  # 이재인
h.insert(20030513)  # 이지후
h.insert(20020415)  # 이진
h.insert(20040327)  # 이진서
h.insert(20030516)  # 이채린
h.insert(20040415)  # 이희서
h.insert(20050207)  # 임서영
h.insert(20021213)  # 임성민
h.insert(20040318)  # 전민서
h.insert(20041121)  # 정세원
h.insert(20020619)  # 정승주
h.insert(20030802)  # 정윤서
h.insert(20050224)  # 정재윤
h.insert(20030917)  # 정혜민
h.insert(20050119)  # 조승연
h.insert(20040509)  # 조예진
h.insert(20050926)  # 주희선
h.insert(20051008)  # 최가온
h.insert(20050704)  # 최수안
h.insert(20040805)  # 한수진
h.insert(20050621)  # 허재희
h.insert(20040611)  # 홍서연
h.insert(20241282)  # 홍서연
h.insert(20050203)  # 홍지연


birth_map = {
    20241282: ["홍서연"],
    20051230: ["신수민"],
    20051225: ["이서영"],
    20051214: ["강민주"],
    20051202: ["김민경"],
    20051112: ["이서영"],
    20051102: ["배시은"],
    20051031: ["김여원"],
    20051028: ["이서진"],
    20051024: ["서홍빈"],
    20051019: ["김예빈"],
    20051009: ["여지혜"],
    20051008: ["최가온"],
    20051005: ["김나영"],
    20051002: ["김명신"],
    20051001: ["김혜인"],
    20050926: ["주희선"],
    20050923: ["김도경"],
    20050920: ["유채원"],
    20050910: ["김채민"],
    20050712: ["오예준"],
    20050704: ["최수안"],
    20050621: ["허재희"],
    20050620: ["손지원"],
    20050602: ["이원진"],
    20050601: ["이유빈"],
    20050519: ["윤서영"],
    20050517: ["윤혜진"],
    20050509: ["오하민"],
    20050501: ["김혜정"],
    20050422: ["김정민"],
    20050417: ["김주하"],
    20050328: ["오세은"],
    20050325: ["남궁수아"],
    20050324: ["김미배"],
    20050316: ["노은서"],
    20050305: ["이서진"],
    20050224: ["정재윤"],
    20050214: ["민고은"],
    20050207: ["임서영"],
    20050203: ["김소민", "홍지연"],
    20050126: ["신희영"],
    20050123: ["김채린"],
    20050119: ["조승연"],
}


#생일 늦은 10명 출력
count = 0
seen = set()
while count < 10 and h.size() > 0:
    b = h.deleteMax()
    if b in birth_map:
        for name in birth_map[b]:
            if (name, b) not in seen:
                print(f"{name}, {b}")
                seen.add((name, b))
                count += 1
            if count == 10:
                break
```

```python
class KthLargest(object):

    def __init__(self, k, nums):
        self.k = k
        self.nums = nums
        heapq.heapify(self.nums)  # Convert nums into a heap

        while len(self.nums) > k:
            heapq.heappop(self.nums)

    def add(self, val):
        heapq.heappush(self.nums, val)

        if len(self.nums) > self.k:
            heapq.heappop(self.nums)

        return self.nums[0]
```
